```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import seaborn as sns
```

```python
dataset = pd.read_excel("golf_df.xlsx")
dataset.tail()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

#nome do carro
le.fit(dataset.clima)
dataset.clima = le.transform(dataset.clima)

#nome do carro
le.fit(dataset.temp)
dataset.temp = le.transform(dataset.temp)

#nome do carro
le.fit(dataset.umidade)
dataset.umidade = le.transform(dataset.umidade)

#nome do carro
le.fit(dataset.vento)
dataset.vento = le.transform(dataset.vento)


#nome do carro
le.fit(dataset.jogar)
dataset.jogar = le.transform(dataset.jogar)
```

```python
dataset
```

```python
#Selecionando todas as colunas menos a de previsão neste caso a coluna  jogar
x = dataset.iloc[:, :-1]
# variavel  y contem as respostas
y = dataset[["jogar"]]
```

```python
# como a base e pequena para separar em treino e teste foi  x e y  diretamente sem a separação de treino e teste
model = LogisticRegression()
model.fit(x, y)
```

```python
predict = model.predict(x)
```
